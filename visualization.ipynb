### Column: Revenues

```python
revenues = pd.DataFrame(data['Revenue'])
```

```python
revenues
```

```python
def isNaN(string):
    return string != string
```

```python
for i in range(0,3761):
    if isNaN(str(revenues.iloc[i,0])) == False:
        if type(revenues.iloc[i,0]) == str:
            revenues.iloc[i,0] = revenues.iloc[i,0].replace('$','')
```

```python
revenues
```

```python
for i in range(0,3762):
        if 'B' in str(revenues.iloc[i,0]):
            a = revenues.iloc[i,0].replace('B','')
            a = float(a) * 1000000000
            revenues.iloc[i,0] = int(a)
        elif 'M' in str(revenues.iloc[i,0]):
            b = revenues.iloc[i,0].replace('M','')
            b = float(b) * 1000000
            revenues.iloc[i,0] = int(b)
        elif ',' in str(revenues.iloc[i,0]):
            a = revenues.iloc[i,0].split(',')
            revenues.iloc[i,0] = a[0]+a[1]
            revenues.iloc[i,0] = float(revenues.iloc[i,0])
        else:
            continue
```

```python
revenues = symbol.merge(revenues, left_index=True, right_index=True, copy=False)
```

```python
revenues.to_csv('revenues.csv')
```
